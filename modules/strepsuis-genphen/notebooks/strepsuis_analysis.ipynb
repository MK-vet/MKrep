## Step 1: Install Package from GitHub

## Step 2: Upload Data Files

Upload your CSV files using the file upload button.


```python
from google.colab import files
import os

os.makedirs('data', exist_ok=True)
print("Upload your CSV files:")
uploaded = files.upload()

for filename in uploaded.keys():
    os.rename(filename, f'data/{filename}')
    print(f"✓ {filename} uploaded")
```

## Step 3: Run Analysis


```python
from strepsuis_genphen import StrepsuisAnalyzer
import os

os.makedirs('output', exist_ok=True)

analyzer = StrepsuisAnalyzer(
    data_dir='data',
    output_dir='output'
)

print("Running analysis...")
results = analyzer.run()

print("Generating reports...")
analyzer.generate_html_report(results)
analyzer.generate_excel_report(results)
print("✓ Complete!")
```

## Step 4: Download Results


```python
from google.colab import files
import shutil

shutil.make_archive('results', 'zip', 'output')
files.download('results.zip')
print("✓ Download complete!")
```
